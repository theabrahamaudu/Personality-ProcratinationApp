## STMP App

```python
import tkinter as tk
from tkinter import filedialog, Text
import os
```

```python
root = tk.Tk()
root.title('STMP App')


files = []

def loadFile():
    
    for widget in frame.winfo_children():
        widget.destroy()
        
    
    filename = filedialog.askopenfilename(initialdir="/", 
                title = 'Select File', 
                filetypes = (('Exel Files','*.xls'),
                             ('Other Excel Files', '*.xlsx'),
                             ('All Files','*.*')))
    files.append(filename)
    print(filename)
    
    
    for file in files:
        label = tk.Label(frame, text = file, bg = "gray")
        label.pack()

#____________________________________________________________________
def runAnalysis():
    for file in files:
        #import libraries
        import numpy as np
        import pandas as pd
        
        
        go = str (files[0])
        
        #Load the data
        #pd.options.display.max_columns = None
        raw_data = pd.read_csv(go)
        
        ## DISC Test
        #extract disc test info
        disc_test = raw_data.loc[:,'Q1':'Q24']
        disc_test = disc_test.T
        mad = disc_test
        mad = mad.reset_index()
        mad = mad.drop(['index'],axis=1)
        disc_test = mad
        #load cross-reference table
        crt = pd.read_csv('crt.csv')

        dom = []
        inf = []
        ste = []
        con = []



        for i in range(len(raw_data)):
            #Convert info to column data
            #disc_test = disc_test[i]
            #disc_test = disc_test.reset_index()



            #Dominance Test

            crt['D'] == disc_test[i]
            D = sum(crt['D'] == disc_test[i])/len(crt['D'] == disc_test[i])
            D = round(D,2)*24
            D = float(D)
            dom.append(D)

            #Influence Test

            crt['I'] == disc_test[i]
            I = sum(crt['I'] == disc_test[i])/len(crt['I'] == disc_test[i])
            I = round(I,2)*24
            I = float(I)
            inf.append(I)

            #Steadiness Test

            crt['S'] == disc_test[i]
            S = sum(crt['S'] == disc_test[i])/len(crt['S'] == disc_test[i])
            S = round(S,2)*24
            S = float(S)
            ste.append(S)

            #Conscienciousness Test

            crt['C'] == disc_test[i]
            C = sum(crt['C'] == disc_test[i])/len(crt['C'] == disc_test[i])
            C = round(C,2)*24
            C = float(C)
            con.append(C)
        
        ## Procrastination Test
        proc_test = raw_data.iloc[:,1:21]
        proc_test = proc_test.T
        #reset the index of the data
        mad = proc_test
        mad = mad.reset_index()
        mad = mad.drop(['index'],axis=1)
        proc_test = mad
        proc = []
        for i in range(len(raw_data)):
            #proc_test = proc_test.iloc[:,i]


            #Extract revere-keyed questions
            reverse_keyed = proc_test.iloc[[2,3,5,7,10,12,13,14,17,19],i]

            #Format reverse-keyed questions
            formatted_keys_func = abs(reverse_keyed - 6)

            #Add formatted questions to extracted reverse-keyed dataframe
            formatted_keys = reverse_keyed.copy()
            formatted_keys = formatted_keys_func

            #Add formatted questions to procratination values
            proc_test_formatted = proc_test.copy()
            proc_test_formatted = proc_test_formatted.drop([2,3,5,7,10,12,13,14,17,19])
            proc_test_formatted = proc_test_formatted[i].append(formatted_keys)
            proc_test_formatted = proc_test_formatted.sort_index()
            #proc_test_formatted = proc_test_formatted.drop(['index'],axis=1)

            #Compute procrastination score

            proc_test_score = sum(proc_test_formatted)
            proc.append(proc_test_score)
        
        
        ## Summary of Results
        summary_table = pd.DataFrame(columns = ['Dominance','Influence',
                                        'Steadiness','Conscienciousness',
                                       'Procrastination'])
        summary_table['Dominance'] = dom
        summary_table['Influence'] = inf
        summary_table['Steadiness'] = ste
        summary_table['Conscienciousness'] = con
        summary_table['Procrastination'] = proc
        summary_table['Gender'] = raw_data['Gender']
        summary_table['Age Range'] = raw_data['Age Range']
        summary_table['Faculty of Study'] = raw_data['Faculty of study (e.g. Life Sciences, Arts, Engineering, etc.)']
        summary_table['Year of Study'] = raw_data['Year of Study']
        summary_table['CGPA'] = raw_data['Current CGPA']
        summary_table['Email'] = raw_data['Email address (Optional)']
        
        summary_table.to_csv('STMP Analysis App ResultX.csv', index = False)
#____________________________________________________________________

canvas = tk.Canvas(root, height = 400, width = 700, bg = "#263D42")
canvas.pack()

frame = tk.Frame(root, bg = "white")
frame.place(relwidth = 0.8, relheight = 0.8, relx = 0.1, rely = 0.1)

openFile = tk.Button(root, text = "Open File", padx = 10,
                    pady = 5, fg = "white", bg = "#263D42", 
                     command = loadFile)
openFile.pack()

runAnalysis = tk.Button(root, text = "Run Analysis", padx = 10,
                    pady = 5, fg = "white", bg = "#263D42",
                       command = runAnalysis)
runAnalysis.pack()





root.mainloop()
```
